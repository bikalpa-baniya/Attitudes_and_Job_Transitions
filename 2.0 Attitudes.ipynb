```python
def remove_sparse_col(df, var_list, year_list, threshold):
  
  num_missing = 0 
  for i in var_list: num_missing = num_missing + df[i].isna().sum() 
  if num_missing > threshold * len(var_list):
    df = df.drop(columns=var_list)
  output = df
  return output



for i in new_var_name_unqiue:
  sub_var_name = [x for x in new_var_name if x[:-6] == i ]
  sub_year = [x[-4:] for x in sub_var_name ]
  attitudes = remove_sparse_col(attitudes, sub_var_name, sub_year, 5000)

  
```

```python
def add_year(df, var_name, ref_year ,begin_year, end_year):
  for i in range(begin_year, end_year+1):
    added_name = var_name + "__"+str(i)
    ref_name = var_name + "__"+ str(ref_year)
    df[added_name] = df[ref_name]
  output = df.copy()
  return output

def add_begining(df, var_name, ref_year):
  added_name = var_name + "__1979"
  ref_name = var_name + "__"+ ref_year
  df[added_name] = df[ref_name]
  return df.copy()

def add_multiple_ref_year(df, var_name, year_list):
  year_list.sort()
  if year_list[0]!= "1979": 
    df = add_begining(df, var_name, year_list[0])
    year_list.insert(0, "1979")
  for i in range(1,len(year_list)):
    df = add_year(df, var_name, year_list[i-1] , int(year_list[i-1]), int(year_list[i])-1)
    if i == len(year_list)-1:
      #From last element in the list to 2018
      df = add_year(df, var_name, year_list[i] , int(year_list[i]), 2018)

  output = df.copy()
  return output 


def expand_year(df, id_list):
  sub_var_list = [x[:-6] for x in df.columns.tolist() if x not in id_list]
  sub_var_list = list(set(sub_var_list))
  sub_var_list.sort()
  #sub_var_list = sub_var_list[0:10]            
  for i in sub_var_list: 
    sub_var_name = [x for x in df.columns.tolist() if x[:-6] == i ] 
    sub_year = [x[-4:] for x in sub_var_name]
    if len(sub_year) == 1:
      df = add_year(df, i , sub_year[0], 1979, 2018)
    else:
      df = add_multiple_ref_year(df, i , sub_year) 
  
  test_list  = [x for x in df.columns.tolist() if x[:-6] in sub_var_list]
  test_list.sort()
  print(len(test_list))



  output = df 
  #output = test_list
  return output 


attitudes = expand_year(attitudes, id_list)


#attitudes_test
```
